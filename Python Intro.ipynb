#### It has been observed that Make is not properly marked. For Eg: Mercedes, AMG, Mercedes-Benz and Mercedes-AMG all should come under one category but it is marked separtely which does not provide us accurate count result 

```python
labels = cars["Identification.Make"].astype('category').cat.categories.tolist()
sizes = [counts[car] for car in labels]
fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=False, startangle=20)
ax1.axis('equal')
plt.show()
```
